## Q19. Plot a bar graph of Purpose vs Miles(Distance). What can you infer from the plot(2 +2 points)
#### Note: Use the original dataframe without dropping the 'NA' values.
#### Hint:You have to plot total/sum miles per purpose

```python
# Use groupby() to group the miles under purpose and sort it in descending order

total_miles = pd.DataFrame(uber_drives.groupby('PURPOSE*').sum()['MILES*'].sort_values(ascending = False))

# reset index so that purpose and miles will be columns otherwise purpose will be the index
total_miles =  total_miles.reset_index()

# plot the data
sns.barplot(x = total_miles['MILES*'], y = total_miles['PURPOSE*']);
```

**Inference**

So, the most miles have been clocked under meeting category while the least is for Charity
